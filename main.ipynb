```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando os datasets em DataFrames do Pandas
closed_deals_df = pd.read_csv("dataset/olist_closed_deals_dataset.csv")
qualified_leads_df = pd.read_csv("dataset/olist_marketing_qualified_leads_dataset.csv")
```

```python
# Analisando as primeiras linhas do DataFrame de negócios fechados
closed_deals_df.head()
```

```python
# Analisando as primeiras linhas do DataFrame de leads qualificados
qualified_leads_df.head()
```

```python
# Fazendo merge dos dois DataFrames com base na coluna 'mql_id'
merged_df = pd.merge(qualified_leads_df, closed_deals_df, on='mql_id', how='left').reset_index(drop=True)
merged_df.info()
merged_df.head()
```

```python
# Verificando valores nulos no DataFrame mesclado
merged_df.isnull().sum()
```

```python
# Removendo linhas com valores nulos das colunas 'won_date' e 'origin'
cleaned_df = merged_df.dropna(subset=['won_date', 'origin']).reset_index(drop=True)
cleaned_df.info()
cleaned_df.head()
```

```python
# Analisando a quantidade de leads por segmento
segment_counts = cleaned_df['business_type'].value_counts()
print(segment_counts)
```

```python
# Calculando o tempo médio para conversão de lead em negócio fechado

# Convertendo as colunas de data para o formato datetime
cleaned_df["first_contact_date"] = pd.to_datetime(cleaned_df["first_contact_date"])
cleaned_df["won_date"] = pd.to_datetime(cleaned_df["won_date"])
# Calculando a diferença em dias entre a data do primeiro contato e a data de fechamento
cleaned_df["time_to_close"] = (cleaned_df["won_date"] - cleaned_df["first_contact_date"]).dt.days
# Calculando a média do tempo para fechamento
average_time_to_close = cleaned_df["time_to_close"].mean()
print(f"Tempo médio para conversão de lead em negócio fechado: {average_time_to_close:.2f} dias")
```

```python
# Calculando conversões por origem
conversion_by_origin = cleaned_df['origin'].value_counts()
print(conversion_by_origin)
```

```python
# Calculando as conversões por mês
cleaned_df['month'] = cleaned_df['first_contact_date'].dt.to_period('M')
conversions_by_month = cleaned_df['month'].value_counts()
print(conversions_by_month)
```
