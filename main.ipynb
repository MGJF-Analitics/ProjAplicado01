```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando os datasets em DataFrames do Pandas
closed_deals_df = pd.read_csv("dataset/olist_closed_deals_dataset.csv")
qualified_leads_df = pd.read_csv("dataset/olist_marketing_qualified_leads_dataset.csv")
```

```python
# Analisando as primeiras linhas do DataFrame de negócios fechados
closed_deals_df.head()
```

```python
# Analisando as primeiras linhas do DataFrame de leads qualificados
qualified_leads_df.head()
```

```python
# Fazendo merge dos dois DataFrames com base na coluna 'mql_id'
merged_df = pd.merge(qualified_leads_df, closed_deals_df, on='mql_id', how='left').reset_index(drop=True)
merged_df.info()
merged_df.head()
```

```python
# Verificando valores nulos no DataFrame mesclado
merged_df.isnull().sum()
```

```python
# Removendo linhas com valores nulos da coluna 'origin'
cleaned_df = merged_df.dropna(subset=['origin']).reset_index(drop=True)
cleaned_df.info()
cleaned_df.head()
```

```python
# Criando coluna indicadora de conversão e verificando quantos leads foram convertidos
cleaned_df['converted'] = cleaned_df['won_date'].notna()
cleaned_df['converted'].value_counts()
```

```python
# Criando DataFrame apenas com leads convertidos (será utilizado para análises de performance)
converted_leads_df = cleaned_df[cleaned_df['converted'] == True].copy()
converted_leads_df.info()
```

```python
# Renomeando para maior clareza conceitual
all_leads_df = cleaned_df.copy()
```

```python
# Convertendo colunas de data para formato datetime em AMBOS os DataFrames
all_leads_df["first_contact_date"] = pd.to_datetime(all_leads_df["first_contact_date"])
all_leads_df["won_date"] = pd.to_datetime(all_leads_df["won_date"])

converted_leads_df["first_contact_date"] = pd.to_datetime(converted_leads_df["first_contact_date"])
converted_leads_df["won_date"] = pd.to_datetime(converted_leads_df["won_date"])
```

```python
# Calculando tempo de conversão para leads convertidos
converted_leads_df["time_to_close"] = (
    converted_leads_df["won_date"] - converted_leads_df["first_contact_date"]
).dt.days

print(f"Tempo médio de conversão: {converted_leads_df['time_to_close'].mean():.2f} dias")
```

```python
# Calculando taxa de conversão
conversion_rate = (all_leads_df['converted'].sum() / len(all_leads_df) * 100)
```

```python
# Métricas gerais
total_leads = len(all_leads_df)
total_conversoes = all_leads_df['converted'].sum()
total_nao_convertidos = (~all_leads_df['converted']).sum()
taxa_conversao_geral = (total_conversoes / total_leads) * 100
```

```python
# Resumo geral dos dados
print("RESUMO DOS DADOS PREPARADOS")
print(f"\nDataset com todos os leads:")
print(f"Total de leads: {total_leads}")
print(f"Leads convertidos: {total_conversoes:,}")
print(f"Leads não convertidos: {total_nao_convertidos:,}")
print(f"Taxa de conversão geral: {conversion_rate:.2f}%")

print(f"\nDataset com leads convertidos:")
print(f"Total: {len(converted_leads_df):,}")
print(f"Período: {converted_leads_df['first_contact_date'].min().strftime('%Y-%m-%d')} a {converted_leads_df['first_contact_date'].max().strftime('%Y-%m-%d')}")
print(f"Tempo médio de conversão: {converted_leads_df['time_to_close'].mean():.2f} dias")

print(f"QUALIDADE DOS DADOS:")
print(f"Valores nulos em 'origin': 0 (100% preenchido)")
print(f"Registros válidos para análise: {len(all_leads_df):,}")
```

```python
# Total de leads por canal de origem
leads_por_origem = all_leads_df['origin'].value_counts()
print("TOTAL DE LEADS POR ORIGEM")
print(leads_por_origem)
```

```python
# Total de conversões por canal de origem
conversoes_por_origem = all_leads_df[all_leads_df['converted'] == True]['origin'].value_counts()
print("\nCONVERSÕES POR ORIGEM")
print(conversoes_por_origem)
```

```python
# Calculando taxa de conversão por origem
taxa_conversao_origem = (conversoes_por_origem / leads_por_origem * 100).round(2)

  # Criando DataFrame consolidado
funil_origem = pd.DataFrame({
    'Total_Leads': leads_por_origem,
    'Conversoes': conversoes_por_origem,
    'Taxa_Conversao_%': taxa_conversao_origem
}).sort_values('Taxa_Conversao_%', ascending=False)

print("\nANÁLISE DE CONVERSÃO POR ORIGEM")
print(funil_origem)
```

```python
 # Gráfico de barras: Total de Leads vs Conversões por Origem
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Total de Leads por Origem
ax[0].barh(funil_origem.index, funil_origem['Total_Leads'], color='steelblue', edgecolor='black')
ax[0].set_xlabel('Número de Leads', fontsize=12)
ax[0].set_ylabel('Canal de Origem', fontsize=12)
ax[0].set_title('Total de Leads por Canal de Origem', fontsize=14, fontweight='bold')
ax[0].grid(axis='x', alpha=0.3, linestyle='--')

# Subplot 2: Taxa de Conversão por Origem
ax[1].barh(funil_origem.index, funil_origem['Taxa_Conversao_%'], color='coral', edgecolor='black')
ax[1].set_xlabel('Taxa de Conversão (%)', fontsize=12)
ax[1].set_ylabel('Canal de Origem', fontsize=12)
ax[1].set_title('Taxa de Conversão por Canal de Origem', fontsize=14, fontweight='bold')
ax[1].grid(axis='x', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
